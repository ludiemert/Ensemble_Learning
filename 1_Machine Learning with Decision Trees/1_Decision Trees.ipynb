## Decision Trees

```python
from sklearn.datasets import load_iris          # Carrega conjuntos de dados prontos para testes e estudos (ex: dataset Iris) flores, iris
from sklearn.tree import DecisionTreeClassifier # Algoritmo de Árvore de Decisão para classificação
from sklearn import tree                        # Ferramentas para manipular e visualizar árvores de decisão
import matplotlib.pyplot as plt                 # Biblioteca para criação de gráficos e visualizações
import pandas as pd                             # Biblioteca para manipulação e análise de dados em tabelas (DataFrames)
```

```python
# carregar o conjunto de dados
iris = load_iris()
```

```python
# ver ele como dicionario
iris
```

```python
# ver as chaves desse dicionario
iris.keys()
```

```python
# ver classes do target, sao 3 classes => 0,1 e 2
iris["target"]
```

```python
# ver as especies das flore
iris["target_names"]
```

```python
# os dados para ver qual especie que as plantas pertencem
iris["data"]
```

```python
# ver melhor as features, para septa e para septa => largura, comprimento eh essa sequencia => 5.1, 3.5, 1.4, 0.2],
iris["feature_names"]
```

```python
# vamos criar um modelo de decisao usando a arvore de decisao: from sklearn.tree import DecisionTreeClassifier

# definir o x e y =>  todas as linhas e duas ultimas colunas

x = iris.data[:, 2:]
y = iris["target"]


```

```python
# instanciar a classe ( e treinar a classe)
tree_clt = DecisionTreeClassifier()
tree_clt.fit(x, y)
```

```python
# pontos do nosso conjunto de dados

plt.scatter(x[:, 0], x[:,1])
```

```python
# visualizar o DecisionTree
# as inf abaixo contruiu o diagrama

# eh seguindo essa arvore que ele toma suas decisoes

fig, ax = plt.subplots(figsize=(15, 8))
tree.plot_tree(tree_clt)
```

```python
# visualizar o conjunto de dados um do lado do outro

# se vai para a direita (seta) eh falso, vai sweguimentando todos os pontos das features eh assim que toma a decisao deles


fig, ax = plt.subplots(1, 2, figsize=(15, 8))
tree.plot_tree(tree_clt, ax=ax[0])
ax[1].scatter(x[:, 0], x[:, 1])
```

```python
# import a numpy para outro exemplo
import numpy as np
```

```python
# codigo como o algoritmo toma essas decisoes

# Plotting decision regions
x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

fig, ax = plt.subplots(sharex="col", sharey="row", figsize=(10, 8))

Z = tree_clt.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

ax.contourf(xx, yy, Z, alpha=0.4)
ax.scatter(x[:, 0], x[:, 1], c=y, s=20, edgecolor="k")

ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])
ax.set_xlabel(iris["feature_names"][2])
ax.set_ylabel(iris["feature_names"][3])

plt.show()

# Conjunto de dados com as 3 classes
# os pontos sao as tomadas de decisao do modelo
# as 3 classes  sao as separacoes do modelo
# ele vai segmentando os dados
```

```python
# visualizar o conjunto de dados um do lado do outro, com cores diferentes, colocando a a variavel c=y

# se vai para a direita (seta) eh falso, vai sweguimentando todos os pontos das features eh assim que toma a decisao deles


fig, ax = plt.subplots(1, 2, figsize=(15, 8))
tree.plot_tree(tree_clt, ax=ax[0])
ax[1].scatter(x[:, 0], x[:, 1], c=y)


# F => direita
# V => esquerda

# as decisoes sao depth ( profundidade da arvore)
# as cxas sao no ou folha da arvore

#  o sktlearn => lib que usamos para fazer a classificacao ele usa o modelo CARTs e sempre faz cortes bilaterais, ele sempre
# divide nossos dados em dois.
```
